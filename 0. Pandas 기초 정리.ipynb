# 02. Filtering & Sorting

```python
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
```

```python
df.iloc[:, ::2] #df의 짝수번째 컬럼의 데이터만을 출력
df[df.quantity==3] #quantity의 값이 3인 데이터를 추출하여 첫 5행 출력
df[df.quantity==3].reset_index(drop=True) #quantity값이 3인 데이터를 추출하여 index를 0부터 정렬
ans=df[['quantity', 'item_price']] #quantity, item_price 두개 컬럼으로 구성된 새로운 데이터 프레임을 정의
ans=df.iloc[:, [1,4]]
ans=df.loc[:, ['quantity', 'item_price']] 
df['new_price']=df['item_price'].str[1:].astype('float') #item_price 컬럼의 달러표시를 제거하여 float 타입으로 new_price 컬럼에 저장
len(df[df.new_price<=5]) #new_price컬럼의 5이하 값을 가지는 데이터 갯수

df[df.item_name=='Chicken Salad Bowl'].reset_index(drop=True) #Chicken Salad BOwl인 데이터 추출하고 index값 초기화
df[(df.item_name =='Chicken Salad Bowl') & (df.new_price<=9)] #조건에 맞는 데이터 출력
df.sort_values('new_price').reset_index(drop=True) #new_price 컬럼을 오름차순하고 index 초기화
df.sort_values('new_price', ascending=False).reset_index(drop=True) #new_price 컬럼을 내림차순하고 index 초기화
df.loc[df.item_name.str.contains('Chips')] #df의 item_name 컬럼 중 chips를 포함하는 경우의 데이터 출력
df.loc[(df.item_name=='Steak Salad') | (df.item_name=='Bowl')] #해당조건에 맞는 데이터 출력

#item_name를 기준으로 중복행을 첫번째 케이스 남기고 제거
df.loc[(df.item_name=='Steak Salad') | (df.item_name=='Bowl')].drop_duplicates('item_name', keep='first')  #'last' : 마지막행
df.loc[df.item_name=='Izze', 'item_name']='Fizzy Lizzy' #item_name의 값 'izze' -> 'Fizzy Lizzy'로 수정
df.choice_description.isnull().sum() #choice_description값이 Nan인 데이터의 갯수 출력
df.loc[df.choice_description.isnull(), 'choice_description']='NoData' #choice_description 결측값에 NoData를 넣음
df[df.new_price>=df.new_price.mean()] #new_price 값의 평균값 이상의 데이터를 인덱싱
df[df.choice_description.str.contains('Black')] #choice_description 값에 Black이 들어가는 경우 인덱싱
df.loc[-df.choice_description.str.contains('Vegetables')] #choice_description 값에 Vegetables 들어가지 않는 경우 인덱싱
df[df.item_name.str.startswith('N')].head(3) #item_name의 값이 N으로 시작하는 데이터 모두 추출
df[df.item_name.str.len()>=15].head(3) #item_name값의 단어 갯수가 15개 이상인 데이터를 인덱싱
lst=[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98] #new_price값이 lst에 해당하는 경우의 데이터 프레임 출력
df.loc[df.new_price.isin(lst)]
```

# 04. Apply, Map

```python
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv', index_col=0)
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}

df['newIncome']  =df.Income_Category.map(lambda x: dic[x])
df
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)

df['newIncome']
```

```python
df['AgeState']  = df.Customer_Age.map(lambda x: x//10 *10)

df['AgeState'].value_counts().sort_index()
```

```python
df['newEduLevel'] = df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

```python
df['newLimit'] = df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
df['newLimit'].value_counts()
```

```python
def check(x):
    if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
        return 1
    else:
        return 0


df['newState'] = df.apply(check,axis=1)

df['newState'].value_counts()
```

```python
def changeGender(x):
    if x =='M':
        return 'male'
    else:
        return 'female'
df['Gender'] = df.Gender.apply(changeGender)
df['Gender'].value_counts()
```

# 07. Merge, Concat

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

pd.concat([df1,df2])
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]
pd.concat([df3,df4],join='inner') #둘다 포함하고 있는 년도에 대해서만 고려함
```

```python
pd.concat([df3,df4],join='outer').fillna(0) #모든 컬럼을 포함하고 결측치는 0으로 대체
```
