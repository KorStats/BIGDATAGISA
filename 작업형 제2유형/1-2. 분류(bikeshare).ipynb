## 기술통계 조회하기

```python
def DA(data):
    da = data.describe(percentiles=[0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95])
    da = da.T
    df1 = data.isna().sum() # 결측값
    df1.name = 'missing'
    df2 = data.median() # 중앙값
    df2.name = 'median'
    df3 = np.var(data) # 분산
    df3.name = 'variance'
    df4 = data.skew() # 왜도
    df4.name = 'skewness'
    df5 = data.kurtosis() # 첨도
    df5.name = 'kurtosis'
    da = pd.concat([da,df1,df2,df3,df4,df5], axis=1) # 모두 합침
    da['total'] = da['count'] + da['missing'] # 전체 데이터 수
    col_nm = da.columns.tolist()
    order = ['total','count','missing','mean','median','std','variance','skewness','kurtosis','min',
             '5%','10%','25%','50%','75%','90%','95%','max']
    col_nm_new=[]
    for i in order:
        col_nm_new.append(i)
    da = da[col_nm_new]
    da = da.round(2) 
    return da

DA(data[col_num])
```

```python
def DA_cat(data, col_cat):
    DA_cat = pd.DataFrame()
    for i in col_cat:
        a = data[i].value_counts(dropna=False).to_frame().sort_index().rename(columns={i:'count'}).reset_index()
        a['col_nm'] = i
        a = a.rename(columns = {'index':'class'})
        a = a[['col_nm','class','count']]
        b=data[i].value_counts(dropna = False, normalize = True).to_frame().sort_index().rename(
        columns = {i:'ratio'}).reset_index()
        b = b['ratio'].to_frame()
        b['ratio'] = b['ratio'].round(2)
        c = pd.concat([a,b], axis = 1)
        DA_cat = pd.concat([DA_cat, c], axis=0)
    DA_cat = DA_cat.reset_index(drop=True)
    return DA_cat

DA2 = DA_cat(data,col_cat+col_bool)
DA2
```

### 이상치 처리 

```python
q1=data['humidity'].quantile(.25) ; q3=data['humidity'].quantile(.75)
iqr=q3-q1
min_iqr=q1-1.5*iqr ; max_iqr=q3+1.5*iqr
min_from_all=data['humidity'].min()
max_from_all=data['humidity'].max()
if(min_iqr<min_from_all):
    min_iqr=min_from_all
if(max_iqr>max_from_all):
    max_iqr=max_from_all
data[(data['humidity']<min_iqr) | (data['humidity']>max_iqr)]
```

```python
data.loc[(data['humidity']<min_iqr), 'humidity']=min_iqr #이상치 보정: 하한치로 보정
data.loc[(data['humidity']>max_iqr), 'humidity']=max_iqr #이상치 보정: 상한치로 보정
data.head(-3)
```

### 날짜데이터 처리 

```python
def dataPreprocessing(df2):
    df=df2.copy()
    df['datetime']=pd.to_datetime(df['datetime'])
    df['year']=df['datetime'].dt.year
    df['month']=df['datetime'].dt.month
    df['day']=df['datetime'].dt.day
    df['weekend']=df['datetime'].dt.weekday
    df['hour']=df['datetime'].dt.hour
    df['minute']=df['datetime'].dt.minute
    return df

pre_data=dataPreprocessing(data).drop(['datetime'], axis=1)
pre_data.head(3)
```

### 범주형 변수 더미화 

```python
pre_data_1=pre_data.copy()
data_dum=pd.get_dummies(pre_data_1)
data_dum.head(10)
```

### 종속변수 전처리(이항형태로 변환 / 4개 클래스로 변환) 

```python
data_dum.loc[data_dum['count']<=150, 'y']=0
data_dum.loc[(data_dum['count']>150) & (data_dum['count']<=300), 'y']=1
data_dum.loc[(data_dum['count']>300) & (data_dum['count']<=450), 'y']=2
data_dum.loc[data_dum['count']>450, 'y']=3
```

## 모델링 

```python
from sklearn.linear_model import LogisticRegression #로지스틱회귀
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.tree import DecisionTreeClassifier #의사결정트리
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis #LDA
from sklearn.naive_bayes import GaussianNB #가우시안NB
from sklearn.ensemble import VotingClassifier, RandomForestClassifier, GradientBoostingClassifier #보팅, 랜덤포레스트, GBM
from xgboost import XGBClassifier #XGB
from sklearn.neural_network import MLPClassifier #MLP
from sklearn.svm import SVC #SVM

lr_clf=LogisticRegression()
knn_clf=KNeighborsClassifier(n_neighbors=8)
dt_clf=DecisionTreeClassifier(random_state=156)
lda_clf=LinearDiscriminantAnalysis(n_components=2)
gnb_clf=GaussianNB()
vo_clf=VotingClassifier(estimators=[('LR', lr_clf), ('KNN', knn_clf)], voting='soft')
rf_clf=RandomForestClassifier(random_state=0, n_jobs=-1)
gb_clf=GradientBoostingClassifier(random_state=0)
xgb_clf=XGBClassifier(n_estimators=400, learning_rate=0.1, max_depth=3)
mlp_clf=MLPClassifier(random_state=1, max_iter=300)
svc_clf=SVC(kernel='rbf')

lr_clf.fit(X_train, y_train)
knn_clf.fit(X_train, y_train)
dt_clf.fit(X_train, y_train)
lda_clf.fit(X_train, y_train)
gnb_clf.fit(X_train, y_train)
vo_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
gb_clf.fit(X_train, y_train)
xgb_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_test, y_test)], verbose=False)
mlp_clf.fit(X_train, y_train)
svc_clf.fit(X_train, y_train)

lr_pred=lr_clf.predict(X_test)
knn_pred=knn_clf.predict(X_test)
dt_pred=dt_clf.predict(X_test)
lda_pred=lda_clf.predict(X_test)
gnb_pred=gnb_clf.predict(X_test)
vo_pred=vo_clf.predict(X_test)
rf_pred=rf_clf.predict(X_test)
gb_pred=gb_clf.predict(X_test)
xgb_pred=xgb_clf.predict(X_test)
mlp_pred=mlp_clf.predict(X_test)
svc_pred=svc_clf.predict(X_test)
```
