## DataFrame 생성하기

```python
import pandas as pd
data={'name':['Jason', 'Molly', np.nan, 'Jake', 'Amy'], 'year': [2012, 2012, 2013, 2014, 2014], 'reports': [4, 24, 31, 2, np.nan]}
df=pd.DataFrame(data, index=['cochice', 'Pima', 'SantaCruz', 'Maricopa', 'Yuma'])
print('#Original DataFrame: \n', df)
df=df.rename(columns={'reports':'new_reports'}) #컬럼 이름 개별 변경
print('#Modify DataFrame: \n', df)
```

## DataFrame 조회하기

```python
print('#변수명 출력:', df.columns)
print('#Unique:', df['year'].unique())
print('#위치기반 iloc: \n', df.iloc[1:4]) 
print('#명칭기반 loc: \n', df.loc[:, 'name'])
print('#명칭기반(조건) \n', df.loc[df['year']==2012, ['name', 'new_reports']])
print('#조건 DataFrame 출력: \n', df[(df['new_reports']>20)]) #df는 온전히 보존
```

## DataFrame 정렬하기

```python
print('## sort_index(axis=0): \n', df.sort_index(axis=0)) #인덱스 정렬
print('## sort_valuesㅣ \n', df.sort_values(by=['year', 'new_reports'], ascending=[False, True]))
print('## value_counts()\n', df['year'].value_counts() ) #빈도분석
print('## isin \n', df['name'].isin(['Amy'])) # 값이 있으면 True 반환
print('## reset_index(): \n', df.reset_index()) #index값 없어지고 index라는 컬럼 생성
print('## set_index: \n', df.set_index('name')) #name 컬럼이 index가 됨
```

## Dummy화

```python
rank_data={'rank':[3,5,1,2,4]}
df_rank=pd.DataFrame(rank_data, index=['cochice', 'Yuma', 'SantaCruz', 'Pima', 'Maricopa'])
df_new=pd.concat([df, df_rank], axis=1) #DataFrame 합치기
df_new
```

## 결측치 처리

```python
df1=df.dropna()    #nan이 하나라도 들어간 행은 삭제    *참고: df.dropna(how='all') : 모두 non인 행만 삭제
df2=df.fillna(0) #결측값을 0으로 채움
df['new_reports']=df['new_reports'].fillna(df['new_reports'].mean()) #결측값을 평균값으로 대체
print('#df1 \n', df1)
print('#df2 \n', df2)
print('#df3 \n', df)
```

```python
df.duplicated() #중복여부 boolean 형태로 반환
df.drop_duplicates() #행 내용이 동일한 경우 제거
df.drop_duplicates(['year'], keep='last') #year 변수 기준 중복값 중 마지막만 남겨둠
```

## 이상치 처리

```python
data=pd.read_csv('D:\Machine Learning\ADP-master\data/bike_marketing.csv') #encoing='cp949', 'utf-8', 'cp949', 'euc-kr'

q1=data['google_adwords'].quantile(.25) ; q3=data['google_adwords'].quantile(.75)
iqr=q3-q1
min_iqr=q1-1.5*iqr ; max_iqr=q3+1.5*iqr
min_from_all=data['google_adwords'].min()
max_from_all=data['google_adwords'].max()
if(min_iqr<min_from_all):
    min_iqr=min_from_all
if(max_iqr>max_from_all):
    max_iqr=max_from_all
data[(data['google_adwords']<min_iqr) | (data['google_adwords']>max_iqr)]
```

```python
data.loc[(data['google_adwords']<min_iqr), 'google_adwords']=min_iqr #이상치 보정: 하한치로 보정
data.loc[(data['google_adwords']>max_iqr), 'google_adwords']=max_iqr #이상치 보정: 상한치로 보정
data.head(-3)
```

```python
data=pd.read_csv('D:\Machine Learning\ADP-master\data/bikeshare.csv')
data.head(10)
```

```python
import datetime

data['datetime'] = pd.to_datetime(data['datetime']) 
data['year'] = data['datetime'].map(lambda x: x.year) # 연
data['month'] = data['datetime'].map(lambda x: x.month) # 월
data['day'] = data['datetime'].map(lambda x: x.day) # 일
data['hour'] = data['datetime'].map(lambda x: x.hour) # 시
data['dayofweek'] = data['datetime'].map(lambda x: x.dayofweek) # 요일
```

```python
data = data.groupby(['year','month','day','hour'])['temp','humidity', 'windspeed','count'].mean().reset_index()
data
```

```python
data['new_date']=data['year'].astype('str')+'-'+data['month'].astype('str')+'-'+data['day'].astype('str')+'-'+data['hour'].astype('str')
data['new_date'] = pd.to_datetime(data['new_date'], format='%Y-%m-%d-%H')  # date 컬럼 형식 변경 -> datetime
data
```
